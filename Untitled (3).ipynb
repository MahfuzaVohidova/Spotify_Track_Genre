# Create Quartiles

```python
import pandas as pd

numeric_columns = dataframe.select_dtypes(include=['number'])

Q1_values = numeric_columns.quantile(q=0.25, axis=0, interpolation='linear')
Q3_values = numeric_columns.quantile(q=0.75, axis=0, interpolation='linear')
IQR = Q3_values - Q1_values

print(IQR)
```

# Identify Outliers

```python
Lower_Limit = Q1_values - 1.5 * IQR 
Upper_Limit = Q3_values + 1.5 * IQR

Outliers =dataframe[((dataframe[IQR.index] < Lower_Limit) | ((dataframe[IQR.index] > Upper_Limit))).any(axis=1)]
print(Outliers)

Percentage = (1 - Outliers.shape[0] / dataframe.shape[0]) * 100
print(Percentage)
```

# Create a box plot without outliers

```python
no_outliers = dataframe[~((dataframe[IQR.index] < Lower_Limit) | (dataframe[IQR.index] > Upper_Limit)).any(axis=1)]
print(no_outliers)

plt.figure(figsize=(6, 3))
no_outliers[no_outliers['explicit'] == 1].boxplot()
plt.title("Boxplot for Explicit = 1 (False)")
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(6, 3))
no_outliers[no_outliers['explicit'] == 0].boxplot()
plt.title("Boxplot for Explicit = 0 (True)")
plt.xticks(rotation=90)
plt.show()


```

# Use OLS Regression

```python

dataframe = pd.read_csv("Spotify.csv")

dataframe['explicit'] = dataframe['explicit'].astype(int)

list_of_columns_to_keep = ['duration_ms', 'explicit', 'danceability',
                            'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness',
                            'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature']

dependent_variable = dataframe["popularity"]

independent_variables = dataframe[list_of_columns_to_keep]

independent_variables = sm.add_constant(independent_variables)

regression_model = sm.OLS(dependent_variable, independent_variables).fit()

print(regression_model.summary())
```

```python

dataframe['explicit'] = dataframe['explicit'].astype(int)

list_of_columns_to_keep = ['duration_ms', 'explicit', 'danceability',
                            'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness',
                            'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature']


dependent_variables_stdev = dataframe["popularity"].astype(float)


independent_variables_stdev = dataframe[list_of_columns_to_keep]


independent_variables_stdev = sm.add_constant(independent_variables_stdev)


regression_model_stdev = sm.OLS(dependent_variables_stdev, independent_variables_stdev).fit()

print(regression_model_stdev.summary())
```

# Analysis of Coefficients

```python
regression_model_stdev_bar = regression_model_stdev.params.sort_values(ascending=True)
```

```python
regression_model_stdev_bar.plot(kind="bar",title="Regressive Coefficients for Popularity")
```

# Validate Predictions

```python
dataframe.columns
```

```python
dataframe_genre = pd.read_csv("Spotify.csv")


dependent_variable_genre = 'popularity'
independent_variables_genre = dataframe_genre[['duration_ms', 'danceability', 'energy', 'loudness',
                                               'mode', 'speechiness', 'acousticness', 'instrumentalness', 
                                               'liveness', 'valence', 'tempo', 'time_signature']]

regression_model_stdev = sm.OLS(dataframe_genre[dependent_variable_genre].astype(float), 
                                sm.add_constant(independent_variables_genre)).fit()


dataframe_genre['Prediction'] = regression_model_stdev.predict()


fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(dataframe_genre['popularity'], label='Popularity', linewidth=2)
ax1.plot(dataframe_genre['duration_ms'], label='Duration', linewidth=2)
ax1.plot(dataframe_genre['danceability'], label='Danceability', linewidth=2)
ax1.plot(dataframe_genre['energy'], label='Energy', linewidth=2)
ax1.plot(dataframe_genre['loudness'], label='Loudness', linewidth=2)
ax1.plot(dataframe_genre['Prediction'], color='green', label='Prediction', linewidth=2)
ax1.set_ylabel('Values')

ax1.set_xlabel('Observation Index')
ax1.set_ylabel('Values')

ax1.legend(loc='upper left')

ax2 = ax1.twinx()

plt.title("Regression Results for Spotify Track Genre")

plt.show()
```

```python
 The prediction model confirms that song duration is a key factor in determining popularity. Overall, song length seems to influence how well a song does on Spotify.
```
